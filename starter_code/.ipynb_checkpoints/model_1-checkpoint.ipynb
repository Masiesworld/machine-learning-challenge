```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co','koi_fpflag_ec',
        'koi_period', 'koi_time0bk', 'koi_impact',
       'koi_duration', 'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol',
       'koi_model_snr', 'koi_slogg', 'koi_srad', 'koi_kepmag']]
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
X = selected_features
y = df['koi_disposition']
```

## Label Encoding & dummy coding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# Check the labels
label_encoder.fit(y)
label_encoder.classes_ ##['CANDIDATE'--0, 'CONFIRMED'--1, 'FALSE POSITIVE'--2]

# transform
y_num = label_encoder.transform(y)
```

```python
y_num = pd.Series(y_num)
print(type(y)) 
print(type(y_num)) 
y_num
```

```python
print(X.shape, y_num.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_num, random_state=1)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)

# Transform
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Model
### Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
score = classifier.score(X_test, y_test)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
```

## Confusion Matrix, Classification Report 

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predictions)
print(cm)
```

```python
plt.figure(figsize=(9,9))
sn.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'crest');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {:.3f}'.format(score)
plt.title(all_sample_title, size = 15);
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
